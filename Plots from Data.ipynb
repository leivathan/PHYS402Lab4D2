```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
singleslitcurrent = np.genfromtxt('SingleSlitrun2_04mm_current.txt',skip_header = 22,skip_footer=5)
singleslitvoltage = np.genfromtxt('SingleSlitrun2_04mm_voltage.txt',skip_header = 22,skip_footer=5)
singleslitlogcurrent = np.genfromtxt('SingleSlitrun2_04mm_logcurrent.txt',skip_header = 22,skip_footer=5)
doubleslitcurrent = np.genfromtxt('DoubleSlitrun3_04mmwidth_5mmspace_current.txt',skip_header = 22,skip_footer=5)
doubleslitvoltage = np.genfromtxt('DoubleSlitrun3_04mmwidth_5mmspace_voltage.txt',skip_header = 22,skip_footer=5)
doubleslitlogcurrent = np.genfromtxt('DoubleSlitrun3_04mmwidth_5mmspace_logcurrent.txt',skip_header = 22,skip_footer=5)
wiresim = np.genfromtxt('Wire_simulation.txt',skip_header=15)
planesim = np.genfromtxt('SemiPlane_simulation.txt',skip_header=3014,skip_footer=2999)
semiplanecurrent = np.genfromtxt('SemiPlanerun5_04mmslitwidth_current.txt',skip_header = 22,skip_footer=5)
semiplanevoltage = np.genfromtxt('SemiPlanerun5_04mmslitwidth_voltage.txt',skip_header = 22,skip_footer=5)
semiplanelogcurrent = np.genfromtxt('SemiPlanerun5_04mmslitwidth_logcurrent.txt',skip_header = 22,skip_footer=5)
wirecurrents = np.genfromtxt('Wirerun6_04mmwidth_27mmgauge_current.txt',skip_header = 22,skip_footer=5)
wirevoltage = np.genfromtxt('Wirerun6_04mmwidth_27mmgauge_voltage.txt',skip_header = 22,skip_footer=5)
wirelogcurrent = np.genfromtxt('Wirerun6_04mmwidth_27mmgauge_logcurrent.txt',skip_header = 22,skip_footer=5)
```

```python
doubxpos,doubcurrent,doubvoltage,doublogcur = doubleslitcurrent[:,0],doubleslitcurrent[:,1],doubleslitcurrent[:,2],np.log10(doubleslitcurrent[:,1])
singxpos,singcurrent,singvoltage,singlogcur = singleslitcurrent[:,0],singleslitcurrent[:,1],singleslitcurrent[:,2],np.log10(singleslitcurrent[:,1])
semixpos,semicurrent,semivoltage,semilogcur = semiplanecurrent[:,0],semiplanecurrent[:,1],semiplanecurrent[:,2],np.log10(semiplanecurrent[:,1])
wirexpos,wirecurrent,wirevoltage,wirelogcur = wirecurrents[:,0],wirecurrents[:,1],wirecurrents[:,2],np.log10(wirecurrents[:,1])
wiresimxpos,wiresimint = wiresim[:,0],wiresim[:,1]
semisimxpos,semisimint = planesim[:,0],planesim[:,1]
```

```python
plt.plot(doubxpos,doubcurrent)
plt.title('Double Slit Diffraction Pattern (Current)')
plt.ylabel('Current (Intensity)')
plt.xlabel('Detector Position (cm)')
print(np.nanmin(doubcurrent))
print(doubxpos[np.where(doubcurrent == np.nanmin(doubcurrent))])
```

```python
plt.plot(doubxpos,doubvoltage)
plt.title('Double Slit Diffraction Pattern (Voltage)')
plt.ylabel('Voltage (Intensity)')
plt.xlabel('Detector Position (cm)')
```

```python
plt.plot(doubxpos,doublogcur)
plt.title('Double Slit Diffraction Pattern (Logarithm of Current)')
plt.ylabel('log(Current) (Intensity)')
plt.xlabel('Detector Position (cm)')
```
