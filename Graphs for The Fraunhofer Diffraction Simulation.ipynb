```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
#The equation for the Fraunhofer single-slit diffraction pattern
def fraunsingle(x):
    number = 60.925*((np.sin(np.pi*.00004*(np.sin(np.arctan(x/1.0)))/(632.8e-9)))/(np.pi*.00004*(np.sin(np.arctan(x/1.0)))/(632.8e-9)))**2
    return number
```

```python
#Creating the ideal single-slit diffraction graph
position = np.arange(-0.025,0.0251,0.0001)
intens = fraunsingle(position)
plt.plot(position,intens)
plt.title('Single Slit Diffraction Pattern Simulation')
plt.ylabel('Current (Intensity)')
plt.xlabel('Detector Position (m)')
```

```python
print(intens[250])
print(position[250])
print(np.nanmin(intens))
print(position[np.where(intens == np.nanmin(intens))])
```

```python
#The equation for the Fraunhofer double-slit diffraction pattern, It's slightly more intense than before
def fraundouble(x):
    theta = np.arctan(x/1.0)
    alpha = np.pi*0.0005*np.sin(theta)/632.8e-9
    beta = np.pi*0.00004*np.sin(theta)/632.8e-9
    number = 34.802*(np.sin(beta)/beta)**2*(np.sin(2*alpha)/np.sin(alpha))**2
    return number
```

```python
#Creating the ideal double-slit diffraction graph
position = np.arange(-.025,.025,0.0001)
intens = fraundouble(position)
plt.plot(position,intens)
plt.title('Double Slit Diffraction Pattern Simulation')
plt.ylabel('Current (Intensity)')
plt.xlabel('Detector Position (m)')
```

```python
print(intens[250])
print(position[250])
print(np.nanmin(intens))
print(position[np.where(intens == np.nanmin(intens))])
```
